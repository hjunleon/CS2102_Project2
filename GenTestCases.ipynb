```python
random_seed = 20
random.seed(random_seed)
```

```python
def get_random_string(length):
    # choose from all lowercase letter
    letters = string.ascii_letters
    return ''.join(random.choice(letters) for i in range(length))

def get_random_number(length):
    choices = [str(x) for x in range(10)]
    return ''.join(random.choice(choices) for i in range(length))

def get_random_int_to(start = 2, upto = 9):
    return random.randint(start, upto)

def get_random_float_from_to(start = 1, mul = 9):
    return round(random.random() * mul, 2) + start

def random_plus_minus(num, mag = 5):
    return num + random.random() * 5 - random.random() * 5 

def get_random_price():
    return round(random.random() * 100, 2)

def random_date():
    d = random.randint(1, int(time.time()))
    return datetime.fromtimestamp(d).strftime('%Y-%m-%d')
def random_timestamp():
    d = random.randint(1, int(time.time()))
    return datetime.fromtimestamp(d)#.strftime('%Y-%m-%d')
```

```python
def is_cancelled(status):
    return status == 'cancelled'

def is_unsuccessful(status):
    return status == 'unsuccessful'
```

```python
NUM_EMPLOYEES = 50
NUM_FACILITIES = 10
NUM_CUSTOMERS = 1000
NUM_DEL_REQ = NUM_CUSTOMERS * 2
NUM_PACKAGES = math.floor(NUM_DEL_REQ * 1.5)
NUM_LEGS = NUM_DEL_REQ * 3
NUM_UNSUCCESSFUL_PICKUP = NUM_DEL_REQ // 100 + 5
# class delivery_status(Enum):
#     submitted = 1
#     evaluated = 2
#     withdrawn = 3
#     accepted = 4
#     completed = 5
#     cancelled = 6
#     unsuccessful = 7
    
all_d_status = ['submitted', 'evaluated', 'withdrawn', 'accepted', 'completed', 'cancelled', 'unsuccessful']
accepted_statuses = set(['accepted', 'completed', 'cancelled', 'unsuccessful'])

```

```python
employee = {}
delivery_staff_id = set()
for i in range(NUM_EMPLOYEES):
    this_id = i + 1
    to_add = {}
    to_add['gender'] = random.choice( ['male', 'female'] )
    to_add['name'] = get_random_string(16)
    to_add["dob"] = random_date()
    to_add["title"] = random.choice(['Mr', 'Ser']) if to_add['gender'] == 'male' else random.choice(['Mrs', 'Miss', "Mdm"])
    to_add["salary"] = round(random.random() * 15000, 2)
    to_add["is_delivery"] = False
    employee[this_id] = to_add
for i in range(min(NUM_EMPLOYEES,10)):
    this_id = i + 1
    employee[this_id]["is_delivery"] = True
    delivery_staff_id.add(i)
employee
```

```python
customers = {}
for i in range(NUM_CUSTOMERS):
    this_id = i + 1
    to_add = {}
    to_add['name'] = get_random_string(16)
    to_add['gender'] = random.choice( ['male', 'female'] )
    to_add['mobile'] = get_random_number(8)
    customers[this_id] = to_add
customers
```

```python
delivery_requests = {}
accepted_req_id = set()
cancelled_or_unsuccessful_requests_id = set()
for i in range(NUM_DEL_REQ):
    this_id = i + 1
    to_add = {}
    to_add['recipient_name'] = get_random_string(16)
    to_add['recipient_addr'] = get_random_string(32)
    to_add['recipient_postal'] = get_random_number(6)
    to_add['pickup_addr'] = get_random_string(32)
    to_add['pickup_postal'] = get_random_number(6)
    to_add['pickup_date'] = random_date()
    to_add['submission_time'] = random_timestamp()
    to_add['num_days_needed'] = get_random_int_to(2, 16)
    to_add['price'] = get_random_price()
    
    to_add['customer_id'] = random.choice(list(customers.keys()))
    to_add['evaluater_id'] = random.choice(list(employee.keys()))
    to_add['status'] = random.choice(all_d_status)
    
    if (to_add['status'] in accepted_statuses):
        to_add['card_number'] = get_random_number(12)
        to_add['payment_time'] = random_timestamp()
        to_add['monitor_id'] = random.choice(list(employee.keys()))
        accepted_req_id.add(this_id)
        
    if (is_cancelled(to_add['status'])):
        to_add['cancel_time'] = random_timestamp()
        cancelled_or_unsuccessful_requests_id.add(this_id)
    
    if (is_unsuccessful(to_add['status'])):
        cancelled_or_unsuccessful_requests_id.add(this_id)
    
    delivery_requests[this_id] = to_add
    
delivery_requests
```

```python
packages = {}
for i in range(NUM_PACKAGES):
    this_id = i + 1
    to_add = {}
    to_add['request_id'] = random.choice(list(delivery_requests.keys()))
    to_add['package_id'] = this_id
    this_id = (to_add['request_id'], to_add['package_id'])
#     print(this_id)
    to_add['reported_height'] = get_random_float_from_to(2, 1000)
    to_add['reported_width'] = get_random_float_from_to(2, 1000)
    to_add['reported_depth'] = get_random_float_from_to(2, 1000)
    to_add['reported_weight'] = get_random_float_from_to(2, 1000)
    to_add['actual_height'] = random_plus_minus(to_add['reported_height'])
    to_add['actual_width'] = random_plus_minus(to_add['reported_width'])
    to_add['actual_depth'] = random_plus_minus(to_add['reported_depth'])
    to_add['actual_weight'] = random_plus_minus(to_add['reported_weight'])
    to_add['content'] = get_random_string(12)
    to_add['estimated_value'] = get_random_price()
    
    
    packages[this_id] = to_add
    
# packages
```

```python
facilities = {}

for i in range(NUM_FACILITIES):
    this_id = i + 1
    to_add = {}
    to_add['address'] = get_random_string(32)
    to_add['postal'] = get_random_string(6)
    facilities[this_id] = to_add
    
facilities
```

```python
failed_pickups = {}

for i in range(NUM_UNSUCCESSFUL_PICKUP):
    to_add = {}
    to_add["reason"] =  get_random_string(32)
    to_add['request_id'] = random.choice(list(delivery_requests.keys()))
    to_add['handler_id'] = random.choice(list(delivery_staff_id))
    to_add['pickup_id'] = i
    this_id = (to_add['request_id'], to_add['pickup_id'])
    # Double check if this needs to be strictly after any other time
    to_add["pickup_time"] = random_timestamp()
    
    failed_pickups[this_id] = to_add
    
failed_pickups
```

```python
legs = {}
# NUM_LEGS
    
for accepted_id in accepted_req_id:
#     print(f"accepted_id: {accepted_id}")
    if (random.choice([True,False])):
        current_timestamp = random_timestamp()
        for leg_id in range(get_random_int_to(0, 6)):
            this_id = (accepted_id, leg_id)
            # Check if need to match the pickup handler's id
            to_add['request_id'] = accepted_id
            to_add['leg_id'] = leg_id
            to_add['handler_id'] = random.choice(list(delivery_staff_id))
            to_add['start_time'] = current_timestamp
            current_timestamp += timedelta(hours=get_random_int_to(1,12))
            to_add['end_time'] = current_timestamp
            legs[this_id] = to_add
            
legs
```

```python
unsuccessful_deliveries = {}
fail_cnt = 5

for k, v in legs.items():
    to_ret = {}
    to_ret['request_id'] = k[0]
    to_ret['leg_id'] = k[1]
    to_ret["reason"] = get_random_string(32)
    
    # Are there any constraints?
    to_ret["attempt_time"] = v['start_time'] + timedelta(hours=get_random_int_to(1,12))
    
    unsuccessful_deliveries[k] = to_ret
    if fail_cnt == 0:
        break
    fail_cnt -= 1

unsuccessful_deliveries
```

```python
return_legs = {}
for request_id in cancelled_or_unsuccessful_requests_id:
    if (random.choice([True,False])):
        current_timestamp = random_timestamp()
        for leg_id in range(get_random_int_to(0, 6)):
            this_id = (request_id, leg_id)
            # Check if need to match the pickup handler's id
            to_add['request_id'] = request_id
            to_add['leg_id'] = leg_id
            to_add['handler_id'] = random.choice(list(delivery_staff_id))
            to_add['start_time'] = current_timestamp
            current_timestamp += timedelta(hours=get_random_int_to(1,12))
            to_add['end_time'] = current_timestamp
            to_add['source_facility'] = random.choice(list(facilities.keys()))
            return_legs[this_id] = to_add
            
return_legs
```

```python
unsuccessful_return_deliveries = {}

for k,v in return_legs.items():
    to_add={}
    to_add['request_id'] = k[0]
    to_add['leg_id'] = k[1]
    to_ret["reason"] = get_random_string(32)
    
    # Are there any constraints?
    to_ret["attempt_time"] = v['start_time'] + timedelta(hours=get_random_int_to(1,12))
    
    
    unsuccessful_return_deliveries[k] = to_add

unsuccessful_return_deliveries
```

```python
# Insert these shit

def insert_table(query, t_name, record):
    try:
        conn = connect__to_database()
        cur = conn.cursor()
        cur.execute(query, record)
        conn.commit()
        count = cur.rowcount
        print(f"{count} record(s) inserted successfully into {t_name} table")
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
    finally:
        cur.close()
        if conn is not None:
            conn.close()
def insert_M_table(query, t_name, recordS):
    has_error = False
    try:
        conn = connect__to_database()
        cur = conn.cursor()
        cur.executemany(query, recordS)
        conn.commit()
        count = cur.rowcount
        print(f"{count} record(s) inserted successfully into {t_name} table")
    except (Exception, psycopg2.DatabaseError) as error:
        print(f"ERROR WHILE INSERTING INTO {t_name}: {error}")
        has_error = True
    finally:
        cur.close()
        if conn is not None:
            conn.close()
        return has_error
# 0: {'name': 'QlIDmJDlCOKjgvLf', 'gender': 'M', 'mobile': '12942917'}
def insert_customers(rows):
    rows = [(row['name'], row['gender'], row['mobile']) for row in rows]
    sql = """INSERT INTO customers(name, gender, mobile) VALUES(%s,%s,%s)"""
#     record_to_insert = (row['name'], row['gender'], row['mobile'])
    return insert_M_table(sql, "customers", rows)
"""
10: {'gender': 'female',
  'name': 'sMgacnVnNzSMfLxY',
  'dob': '1989-04-04',
  'title': 'Miss',
  'salary': 748.21,
  'is_delivery': False}
"""
#     cols = row.keys()
#     vals = row.values()
def insert_employees(rows):
    rows = [(row['name'], row['gender'], row['dob'], row['title'], row['salary']) for row in rows]
    sql = """INSERT INTO employees(name, gender, dob, title, salary) VALUES(%s,%s,%s,%s,%s)"""
#     record_to_insert = (row['name'], row['gender'], row['dob'], row['title'], row['salary'])
    return insert_M_table(sql, "employees", rows)
    

def insert_delivery_staff(ids):
    rows = [(str(employee_id),) for employee_id in ids]
    sql = """INSERT INTO delivery_staff(id) VALUES(%s)"""
#     record_to_insert = (str(employee_id))
    return insert_M_table(sql, "delivery_staff", rows)
"""
{'recipient_name': 'kChQvfBAPYDfORAw',
 'recipient_addr': 'vBgAfZyOsPHEDXqwrYRWnznFccGOgvVz',
 'recipient_postal': '326807',
 'pickup_addr': 'TTvakUWHMTcyGDBUmeHrSvyIvOxBrzTJ',
 'pickup_postal': '291330',
 'pickup_date': '1990-12-13',
 'submission_time': datetime.datetime(1994, 9, 16, 6, 26, 12),
 'num_days_needed': 5,
 'price': 85.37,
 'customer_id': 279,
 'evaluater_id': 20,
 'status': 'completed',
 'card_number': '263466782615',
 'payment_time': datetime.datetime(1982, 11, 3, 9, 51, 50),
 'monitor_id': 13}
"""
def insert_delivery_requests(rows):
    rows = [(
        row["customer_id"], 
        row["evaluater_id"], 
        row["status"], 
        row["pickup_addr"], 
        row["pickup_postal"], 
        row["recipient_name"], 
        row["recipient_addr"], 
        row['recipient_postal'], 
        row["submission_time"], 
        row["pickup_date"], 
        row["num_days_needed"], 
        row["price"]
    ) 
            for row in rows]
    sql = """INSERT INTO delivery_requests(customer_id, evaluater_id, status, pickup_addr, pickup_postal, recipient_name, recipient_addr, recipient_postal, submission_time, pickup_date, num_days_needed, price) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
#     record_to_insert = (row["customer_id"], row["evaluater_id"], row["status"], row["pickup_addr"], row["pickup_postal"], row["recipient_name"], row["recipient_addr"], row['recipient_postal'], row["submission_time"], row["pickup_date"], row["num_days_needed"], row["price"])
    return insert_M_table(sql, "delivery_requests", rows)
"""
(798, 0): {'request_id': 798,
  'package_id': 0,
  'reported_height': 602.4,
  'reported_width': 894.45,
  'reported_depth': 731.14,
  'reported_weight': 486.95,
  'actual_height': 604.0399025874523,
  'actual_width': 890.3237229183775,
  'actual_depth': 727.9506293744329,
  'actual_weight': 490.55992195699986,
  'content': 'HqhKjLOwASwJ',
  'estimated_value': 79.37}
"""
def insert_packages(rows):
    sql = """INSERT INTO packages(
            request_id,
            package_id,
            reported_height, 
            reported_width, 
            reported_depth, 
            reported_weight, 
            actual_height, 
            actual_width, 
            actual_depth, 
            actual_weight, 
            content, 
            estimated_value
        ) VALUES(%s,%s, %s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
    rows = [
        (
            row["request_id"], 
            row["package_id"], 
            row["reported_height"], 
            row["reported_width"], 
            row["reported_depth"], 
            row["reported_weight"], 
            row["actual_height"], 
            row["actual_width"], 
            row["actual_depth"], 
            row['actual_weight'], 
            row["content"], 
            row["estimated_value"]
        ) for row in rows
    ]
    return insert_M_table(sql, "packages", rows)

def test_inserts():
    if not have_inserted['customers']:
        have_inserted['customers'] = not insert_customers(customers.values())
    if not have_inserted['employees']:
        have_inserted['employees'] = not insert_employees(employee.values())
    if not have_inserted['d_staff']:
        to_insert = [k for k,v in employee.items() if v['is_delivery']]
        have_inserted['d_staff'] = not insert_delivery_staff(to_insert)
        
    if not have_inserted['d_req']:
        have_inserted['d_req'] = not insert_delivery_requests(delivery_requests.values())
    if not have_inserted['packages']:
        have_inserted['packages'] = not insert_packages(packages.values())
        
```
