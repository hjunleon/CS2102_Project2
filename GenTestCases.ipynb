```python
def get_random_string(length):
    # choose from all lowercase letter
    letters = string.ascii_letters
    return ''.join(random.choice(letters) for i in range(length))

def get_random_number(length):
    choices = [str(x) for x in range(10)]
    return ''.join(random.choice(choices) for i in range(length))

def get_random_int_to(start = 2, upto = 9):
    return random.randint(start, upto)

def get_random_float_from_to(start = 1, mul = 9):
    return round(random.random() * mul, 2) + start

def random_plus_minus(num, mag = 5):
    return num + random.random() * 5 - random.random() * 5 

def get_random_price():
    return round(random.random() * 100, 2)

def random_date():
    d = random.randint(1, int(time.time()))
    return datetime.fromtimestamp(d).strftime('%Y-%m-%d')
def random_timestamp():
    d = random.randint(1, int(time.time()))
    return datetime.fromtimestamp(d)#.strftime('%Y-%m-%d')
```

```python
def is_cancelled(status):
    return status == 'cancelled'

def is_unsuccessful(status):
    return status == 'unsuccessful'
```

```python
NUM_EMPLOYEES = 50
NUM_FACILITIES = 10
NUM_CUSTOMERS = 1000
NUM_DEL_REQ = NUM_CUSTOMERS * 2
NUM_PACKAGES = math.floor(NUM_DEL_REQ * 1.5)
NUM_LEGS = NUM_DEL_REQ * 3
NUM_UNSUCCESSFUL_PICKUP = NUM_DEL_REQ // 100 + 5
# class delivery_status(Enum):
#     submitted = 1
#     evaluated = 2
#     withdrawn = 3
#     accepted = 4
#     completed = 5
#     cancelled = 6
#     unsuccessful = 7
    
all_d_status = ['submitted', 'evaluated', 'withdrawn', 'accepted', 'completed', 'cancelled', 'unsuccessful']
accepted_statuses = set(['accepted', 'completed', 'cancelled', 'unsuccessful'])

```

```python
employee = {}
delivery_staff_id = set()
for i in range(NUM_EMPLOYEES):
    this_id = i + 1
    to_add = {}
    to_add['gender'] = random.choice( ['male', 'female'] )
    to_add['name'] = get_random_string(16)
    to_add["dob"] = random_date()
    to_add["title"] = random.choice(['Mr', 'Ser']) if to_add['gender'] == 'male' else random.choice(['Mrs', 'Miss', "Mdm"])
    to_add["salary"] = round(random.random() * 15000, 2)
    to_add["is_delivery"] = False
    employee[this_id] = to_add
for i in range(min(NUM_EMPLOYEES,10)):
    this_id = i + 1
    employee[this_id]["is_delivery"] = True
    delivery_staff_id.add(this_id)
employee
```

```python
customers = {}
for i in range(NUM_CUSTOMERS):
    this_id = i + 1
    to_add = {}
    to_add['name'] = get_random_string(16)
    to_add['gender'] = random.choice( ['male', 'female'] )
    to_add['mobile'] = get_random_number(8)
    customers[this_id] = to_add
customers
```

```python
delivery_requests = {}
accepted_req_id = set()
cancelled_or_unsuccessful_requests_id = set()
for i in range(NUM_DEL_REQ):
    this_id = i + 1
    to_add = {}
    to_add['recipient_name'] = get_random_string(16)
    to_add['recipient_addr'] = get_random_string(32)
    to_add['recipient_postal'] = get_random_number(6)
    to_add['pickup_addr'] = get_random_string(32)
    to_add['pickup_postal'] = get_random_number(6)
    to_add['pickup_date'] = random_date()
    current_timestamp = random_timestamp()
    to_add['submission_time'] = current_timestamp
    to_add['num_days_needed'] = get_random_int_to(2, 16)
    to_add['price'] = get_random_price()
    
    to_add['customer_id'] = random.choice(list(customers.keys()))
    to_add['evaluater_id'] = random.choice(list(employee.keys()))
    to_add['status'] = random.choice(all_d_status)
    
    if (to_add['status'] in accepted_statuses):
        to_add['card_number'] = get_random_number(12)
        to_add['payment_time'] = random_timestamp()
        to_add['monitor_id'] = random.choice(list(employee.keys()))
        accepted_req_id.add(this_id)
        
    if (is_cancelled(to_add['status'])):
        current_timestamp += timedelta(hours=get_random_int_to(1,12))
        to_add['cancel_time'] = current_timestamp
        cancelled_or_unsuccessful_requests_id.add(this_id)
    
    if (is_unsuccessful(to_add['status'])):
        cancelled_or_unsuccessful_requests_id.add(this_id)
    
    delivery_requests[this_id] = to_add
    
delivery_requests
```

```python
packages = {}
for request_id in delivery_requests.keys():
    current_timestamp = random_timestamp()
    for package_id in range(1, get_random_int_to(2, 6)):
        to_add = {}
        to_add['request_id'] = request_id
        to_add['package_id'] = package_id
        this_id = (to_add['request_id'], to_add['package_id'])
        to_add['reported_height'] = get_random_float_from_to(2, 1000)
        to_add['reported_width'] = get_random_float_from_to(2, 1000)
        to_add['reported_depth'] = get_random_float_from_to(2, 1000)
        to_add['reported_weight'] = get_random_float_from_to(2, 1000)
        to_add['actual_height'] = random_plus_minus(to_add['reported_height'])
        to_add['actual_width'] = random_plus_minus(to_add['reported_width'])
        to_add['actual_depth'] = random_plus_minus(to_add['reported_depth'])
        to_add['actual_weight'] = random_plus_minus(to_add['reported_weight'])
        to_add['content'] = get_random_string(12)
        to_add['estimated_value'] = get_random_price()   
        packages[this_id] = to_add
    
packages
```

```python
failed_pickups = {}
req_to_failed_pickup = {}
for accepted_id in accepted_req_id:
    if (random.choice([True,False])):
        cur_time = delivery_requests[accepted_id]['submission_time']
        last_pickup_id = 1
        for pickup_id in range(1, get_random_int_to(2, 4)):
            to_add = {}
            to_add["reason"] =  get_random_string(32)
            to_add['request_id'] = accepted_id
            to_add['handler_id'] = random.choice(list(delivery_staff_id))
            to_add['pickup_id'] = pickup_id
            this_id = (to_add['request_id'], to_add['pickup_id'])
            # Double check if this needs to be strictly after any other time
            cur_time += timedelta(hours=get_random_int_to(1,12))
            to_add["pickup_time"] = cur_time

            failed_pickups[this_id] = to_add
            last_pickup_id = pickup_id
        
        req_to_failed_pickup[accepted_id] = last_pickup_id
        
failed_pickups
```

```python
legs = {}
# NUM_LEGS
req_to_last_leg = {}
    
    
for accepted_id in accepted_req_id:
#     print(f"accepted_id: {accepted_id}")
    if (random.choice([True,False])):
        cur_time = delivery_requests[accepted_id]['submission_time']
        if accepted_id in req_to_failed_pickup:
            k = (accepted_id, req_to_failed_pickup[accepted_id])
            cur_time = failed_pickups[k]['pickup_time']
        leg_id  = 1
        up_to = get_random_int_to(2, 7)
        while(leg_id < up_to):
            to_add = {}
            this_id = (accepted_id, leg_id)
            # Check if need to match the pickup handler's id
            to_add['request_id'] = accepted_id
            to_add['leg_id'] = leg_id
            to_add['handler_id'] = random.choice(list(delivery_staff_id))
            cur_time += timedelta(hours=get_random_int_to(1,12))
            to_add['start_time'] = cur_time
            cur_time += timedelta(hours=get_random_int_to(1,12))
            to_add['destination_facility'] = random.choice(list(facilities.keys()))
            if leg_id < (up_to - 1):
                to_add['end_time'] = cur_time
            else:
                # None means mid way, a timestamp means complete
                to_add['end_time'] = random.choice([None, cur_time])
            legs[this_id] = to_add
            leg_id += 1
        req_to_last_leg[accepted_id] = up_to - 1 
legs
```

```python
unsuccessful_deliveries = {}
fail_cnt = 10

cur_req = 0
cur_fail_reqs = 0

for k, v in legs.items():
    to_ret = {}
    to_ret['request_id'] = k[0]
    
    to_ret['leg_id'] = k[1]
    if cur_req == to_ret['request_id']:
        cur_fail_reqs += 1
    else:
        if fail_cnt == 0:
            break
        fail_cnt -= 1
        cur_req = to_ret['request_id']
        cur_fail_reqs = 0
    
    if cur_fail_reqs >= 3:
        continue
    
    to_ret["reason"] = get_random_string(32)
    
    # Are there any constraints?
    to_ret["attempt_time"] = v['start_time'] + timedelta(hours=get_random_int_to(1,12))
    
    unsuccessful_deliveries[k] = to_ret
    

unsuccessful_deliveries
```

```python
return_legs = {}
for request_id in cancelled_or_unsuccessful_requests_id:
    if (random.choice([True,False]) and request_id in req_to_last_leg):
        current_timestamp = random_timestamp()
        status = delivery_requests[request_id]['status']
        if (is_cancelled(status)):
            current_timestamp = delivery_requests[request_id]['cancel_time']
            current_timestamp += timedelta(hours=get_random_int_to(1,12))
            
        last_leg = req_to_last_leg[request_id]
#         print(legs[(request_id, last_leg)])
        last_end_time = legs[(request_id, last_leg)]['end_time']
#         print(f"last_end_time: {last_end_time}")
        if last_end_time == None:
            continue
            
        for leg_id in range(1, get_random_int_to(2, 6)):
#             print(f"current_timestamp: {current_timestamp}")
            if (last_end_time >= current_timestamp):
                break
            to_add = {}
            this_id = (request_id, leg_id)
            # Check if need to match the pickup handler's id
            to_add['request_id'] = request_id
            to_add['leg_id'] = leg_id
            to_add['handler_id'] = random.choice(list(delivery_staff_id))
            to_add['start_time'] = current_timestamp
            current_timestamp += timedelta(hours=get_random_int_to(1,12))
            to_add['end_time'] = current_timestamp
            to_add['source_facility'] = random.choice(list(facilities.keys()))
            return_legs[this_id] = to_add
            
return_legs
```

```python
# Insert these shit

def insert_table(query, t_name, record):
    try:
        conn = connect__to_database()
        cur = conn.cursor()
        cur.execute(query, record)
        conn.commit()
        count = cur.rowcount
        print(f"{count} record(s) inserted successfully into {t_name} table")
    except (Exception, psycopg2.DatabaseError) as error:
        print(error)
    finally:
        cur.close()
        if conn is not None:
            conn.close()
def insert_M_table(query, t_name, recordS):
    has_error = False
    try:
        conn = connect__to_database()
        cur = conn.cursor()
        cur.executemany(query, recordS)
        conn.commit()
        count = cur.rowcount
        print(f"{count} record(s) inserted successfully into {t_name} table")
    except (Exception, psycopg2.DatabaseError) as error:
        print(f"ERROR WHILE INSERTING INTO {t_name}: {error}")
        has_error = True
    finally:
        cur.close()
        if conn is not None:
            conn.close()
        return has_error
    
def insert_M_table_no_commit(query, t_name, recordS, conn):
    
    has_error = False
    try:
        cur = conn.cursor()
        cur.executemany(query, recordS)
#         count = cur.rowcount
#         print(f"{count} record(s) inserted successfully into {t_name} table")
    except (Exception, psycopg2.DatabaseError) as error:
        print(f"ERROR WHILE INSERTING INTO {t_name}: {error}")
        has_error = True
#         if conn is not None:
#             conn.close()
    finally:
        cur.close()
        return has_error
    
# 0: {'name': 'QlIDmJDlCOKjgvLf', 'gender': 'M', 'mobile': '12942917'}
def insert_customers(rows):
    rows = [(row['name'], row['gender'], row['mobile']) for row in rows]
    sql = """INSERT INTO customers(name, gender, mobile) VALUES(%s,%s,%s)"""
#     record_to_insert = (row['name'], row['gender'], row['mobile'])
    to_ret = insert_M_table_no_commit(sql, "customers", rows, conn)
    return to_ret
def insert_employees(rows):
    rows = [(row['name'], row['gender'], row['dob'], row['title'], row['salary']) for row in rows]
    sql = """INSERT INTO employees(name, gender, dob, title, salary) VALUES(%s,%s,%s,%s,%s)"""
    to_ret = insert_M_table_no_commit(sql, "employees", rows, conn)
    return to_ret

def insert_delivery_staff(ids):
    rows = [(str(employee_id),) for employee_id in ids]
    sql = """INSERT INTO delivery_staff(id) VALUES(%s)"""
    to_ret = insert_M_table_no_commit(sql, "delivery_staff", rows, conn)
    return to_ret
def insert_delivery_requests(rows):
    rows = [(
        row["customer_id"], 
        row["evaluater_id"], 
        row["status"], 
        row["pickup_addr"], 
        row["pickup_postal"], 
        row["recipient_name"], 
        row["recipient_addr"], 
        row['recipient_postal'], 
        row["submission_time"], 
        row["pickup_date"], 
        row["num_days_needed"], 
        row["price"]
    ) 
            for row in rows]
    sql = """INSERT INTO delivery_requests(customer_id, evaluater_id, status, pickup_addr, pickup_postal, recipient_name, recipient_addr, recipient_postal, submission_time, pickup_date, num_days_needed, price) VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
    to_ret = insert_M_table_no_commit(sql, "delivery_requests", rows, conn)
    return to_ret
def insert_packages(rows):
    sql = """INSERT INTO packages(
            request_id,
            package_id,
            reported_height, 
            reported_width, 
            reported_depth, 
            reported_weight, 
            actual_height, 
            actual_width, 
            actual_depth, 
            actual_weight, 
            content, 
            estimated_value
        ) VALUES(%s,%s, %s,%s,%s,%s,%s,%s,%s,%s,%s,%s)"""
    rows = [
        (
            row["request_id"], 
            row["package_id"], 
            row["reported_height"], 
            row["reported_width"], 
            row["reported_depth"], 
            row["reported_weight"], 
            row["actual_height"], 
            row["actual_width"], 
            row["actual_depth"], 
            row['actual_weight'], 
            row["content"], 
            row["estimated_value"]
        ) for row in rows
    ]
    to_ret = insert_M_table_no_commit(sql, "packages", rows, conn)
    return to_ret

def insert_facilities(rows):
    sql = """INSERT INTO facilities(
            address,
            postal
        ) VALUES(%s,%s)"""
    rows = [
        (
            row["address"], 
            row["postal"]
        ) for row in rows
    ]
    to_ret = insert_M_table_no_commit(sql, "facilities", rows, conn)
    return to_ret

def insert_failed_pickups(rows):
    sql = """INSERT INTO unsuccessful_pickups(
            request_id,
            pickup_id,
            handler_id,
            pickup_time,
            reason
        ) VALUES(%s,%s,%s,%s,%s)"""
    rows = [
        (
            row["request_id"], 
            row["pickup_id"], 
            row["handler_id"], 
            row["pickup_time"], 
            row["reason"]
        ) for row in rows
    ]
    to_ret = insert_M_table_no_commit(sql, "unsuccessful_pickups", rows, conn)
    return to_ret

def insert_legs(rows):
    sql = """INSERT INTO legs(
            request_id,
            leg_id,
            handler_id,
            start_time,
            end_time,
            destination_facility
        ) VALUES(%s,%s,%s,%s,%s,%s)"""
    rows = [
        (
            row["request_id"], 
            str(row["leg_id"]), 
            row["handler_id"], 
            row["start_time"], 
            row["end_time"], 
            row["destination_facility"]
        ) for row in rows
    ]
    to_ret = insert_M_table_no_commit(sql, "legs", rows, conn)
    return to_ret
    
def insert_unsuccessful_deliveries(rows):
    sql = """INSERT INTO unsuccessful_deliveries(
            request_id,
            leg_id,
            attempt_time,
            reason
        ) VALUES(%s,%s,%s,%s)"""
    rows = [
        (
            row["request_id"], 
            row["leg_id"], 
            row["attempt_time"], 
            row["reason"]
        ) for row in rows
    ]
    to_ret = insert_M_table_no_commit(sql, "unsuccessful_deliveries", rows, conn)
    return to_ret
    
def insert_return_legs(rows):
    sql = """INSERT INTO return_legs(
            request_id,
            leg_id,
            handler_id,
            start_time,
            end_time,
            source_facility
        ) VALUES(%s,%s,%s,%s,%s,%s)"""
    rows = [
        (
            str(row["request_id"]), 
            str(row["leg_id"]), 
            str(row["handler_id"]), 
            row["start_time"], 
            row["end_time"], 
            row["source_facility"]
        ) for row in rows
    ]
    to_ret = insert_M_table_no_commit(sql, "legs", rows, conn)
    return to_ret
    
def insert_u_return_deliveries(rows):
    sql = """INSERT INTO unsuccessful_return_deliveries(
            request_id,
            leg_id,
            attempt_time,
            reason
        ) VALUES(%s,%s,%s,%s)"""
    rows = [
        (
            str(row["request_id"]), 
            str(row["leg_id"]), 
            row["attempt_time"], 
            row["reason"]
        ) for row in rows
    ]
    to_ret = insert_M_table_no_commit(sql, "unsuccessful_return_deliveries", rows, conn)
    return to_ret

def insert_accepeted_requests(rows):
    sql = """INSERT INTO accepted_requests(
            id,
            card_number,
            payment_time,
            monitor_id
        ) VALUES(%s,%s,%s,%s)"""
    rows = [
        (
            str(row["id"]), 
            row["card_number"], 
            row["payment_time"], 
            row["monitor_id"]
        ) for row in rows
    ]
    to_ret = insert_M_table_no_commit(sql, "accepted_requests", rows, conn)
    return to_ret
    
def insert_c_or_u_requests(ids):
    sql = """INSERT INTO cancelled_or_unsuccessful_requests(
            id
        ) VALUES(%s)"""
    rows = [
        (
            str(c_u_id),
        ) for c_u_id in ids
    ]
    to_ret = insert_M_table_no_commit(sql, "cancelled_or_unsuccessful_requests", rows, conn)
    return to_ret

def insert_c_requests(rows):
    sql = """INSERT INTO cancelled_requests(
            id,
            cancel_time
        ) VALUES(%s,%s)"""
    rows = [
        (
            str(row["id"]), 
            row["cancel_time"]
        ) for row in rows
    ]
    to_ret = insert_M_table_no_commit(sql, "cancelled_requests", rows, conn)
    return to_ret

def test_inserts():
    if not have_inserted['customers']:
        have_inserted['customers'] = not insert_customers(customers.values())
    if not have_inserted['employees']:
        have_inserted['employees'] = not insert_employees(employee.values())
    if not have_inserted['d_staff']:
        to_insert = [k for k,v in employee.items() if v['is_delivery']]
        have_inserted['d_staff'] = not insert_delivery_staff(to_insert)
        
    if not have_inserted['d_req']:
        have_inserted['d_req'] = not insert_delivery_requests(delivery_requests.values())
    if not have_inserted['accepted_requests']:
        accepted_req_id
        have_inserted['accepted_requests'] = not insert_accepeted_requests([
          {
            "id":k,
            **v
          }  for k,v in delivery_requests.items() if k in accepted_req_id
        ])
    
    if not have_inserted['c_or_u_requests']:
        accepted_req_id
        have_inserted['c_or_u_requests'] = not insert_c_or_u_requests(list(cancelled_or_unsuccessful_requests_id))
        
    if not have_inserted['c_requests']:
        accepted_req_id
        have_inserted['c_requests'] = not insert_c_requests([
          {
            "id":k,
            "cancel_time": v['cancel_time']
          }  for k,v in delivery_requests.items() if 'cancel_time' in v
        ])
    if not have_inserted['packages']:
        have_inserted['packages'] = not insert_packages(packages.values())
    if not have_inserted['facilities']:
        have_inserted['facilities'] = not insert_facilities(facilities.values())
    if not have_inserted['unsuccessful_pickups']:
        have_inserted['unsuccessful_pickups'] = not insert_failed_pickups(failed_pickups.values())
        
    if not have_inserted['legs']:
        have_inserted['legs'] = not insert_legs([
            {
                "leg_id":k,
                **v
            } for k,v in legs.items()
        ])
        
    if not have_inserted['unsuccessful_deliveries']:
        have_inserted['unsuccessful_deliveries'] = not insert_unsuccessful_deliveries(unsuccessful_deliveries.values())
        
    if not have_inserted['return_legs']:
        have_inserted['return_legs'] = not insert_return_legs(return_legs.values())
        
    if not have_inserted['unsuccessful_return_deliveries']:
        have_inserted['unsuccessful_return_deliveries'] = not insert_u_return_deliveries(unsuccessful_return_deliveries.values())

        
    
        
```
